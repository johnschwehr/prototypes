```python
import pandas as pd

pd.set_option('display.max_rows',15)
```

```python
workbook = pd.read_excel('bulk-ad8fvhufsemyg-20210816-20210823-1629761825069.xlsx',sheet_name='Sponsored Products Campaigns')
workbook.head()
```

```python
workbook.columns
```

```python
todrop = ['Record ID','Campaign ID','Portfolio ID','Campaign Start Date','Campaign End Date','Campaign Targeting Type','Ad Group','Keyword or Product Targeting','Product Targeting ID','Match Type','Bidding strategy','Placement Type','Increase bids by placement']
mod_workbook = workbook.drop(labels=todrop,axis=1)
mod_workbook.head()
```

```python
mod_workbook['ACoS'] = mod_workbook['ACoS'].str.rstrip('%').astype('float')
mod_workbook.info()
```

```python
mod_workbook = mod_workbook[(mod_workbook['Record Type']=='Keyword') | (mod_workbook['Record Type']=='Product Targeting')]
mod_workbook
```

```python
mod_workbook = mod_workbook.sort_values(by=['ACoS'],ascending=False)
mod_workbook
```

```python
# Modify max bids according to PPC rules
acos_threshold = [[0,15,0.04],[15,20,0.03],[20,25,0.02],[25,30,0],[30,35,-0.02],[35,40,-0.04],[40,50,-0.06],[50,80,-0.08],[80,100,-0.1],[100,150,-0.12],[150,200,-0.14],[200,None,-0.16]]

for threshold in acos_threshold:
    if threshold[1] is not None:
        mod_workbook.loc[(mod_workbook.ACoS > threshold[0]) & (mod_workbook.ACoS <= threshold[1]), 'Max Bid'] = mod_workbook.loc[(mod_workbook.ACoS > threshold[0]) & (mod_workbook.ACoS <= threshold[1]), 'Max Bid'] + threshold[2]
    else:
        mod_workbook.loc[mod_workbook.ACoS > threshold[0], 'Max Bid'] = mod_workbook.loc[mod_workbook.ACoS > threshold[0], 'Max Bid'] + threshold[2]

mod_workbook
```
